This is to only select 20 generations per initial yield curve generation in order to just have 200 scenarios

Also index is reindexed for use in comparison notebook

```python
import pandas as pd

# Load CSV
df = pd.read_csv("generated_scenarios_from_sampled_curves.csv")

# Keep only scenarios with ID <= 20
filtered_df = df[df['Scenario_ID'] <= 20]

# Drop the last column
filtered_df = filtered_df.iloc[:, :-1]

# Reset index so it's contiguous (0, 1, 2, ...) before reassigning
filtered_df = filtered_df.reset_index(drop=True)

# Reassign Scenario_ID to be unique (1-200) based on row position
filtered_df['Scenario_ID'] = (filtered_df.index // 60) + 1

# Save back to CSV
filtered_df.to_csv("filtered_scenarios_from_sampled_curves 25-4.csv", index=False)
```
