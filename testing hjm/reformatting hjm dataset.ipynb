15-4: The ordering and column names did not align with the one of our generated dataset: this notebook contains to code to quickly reformat.
Going form an ordering on first timestep then on scenarios to first on scenario ID nr then timestep

```python
import pandas as pd

# ── 1. Load ────────────────────────────────────────────────────────────────────
hjm = pd.read_csv("hjm_pca_all_generations.csv") # change to hjm dataset that needs to be reformatted
 
# ── 2. Rename columns to match MVP1 ───────────────────────────────────────────
col_map = {
    "simulation_id": "Scenario_ID",
    "time_step":     "Month",
    "Yield_1M":  "Y_1M",
    "Yield_3M":  "Y_3M",
    "Yield_6M":  "Y_6M",
    "Yield_1Y":  "Y_1Y",
    "Yield_2Y":  "Y_2Y",
    "Yield_3Y":  "Y_3Y",
    "Yield_5Y":  "Y_5Y",
    "Yield_7Y":  "Y_7Y",
    "Yield_10Y": "Y_10Y",
    "Yield_20Y": "Y_20Y",
    "Yield_30Y": "Y_30Y",
}
hjm = hjm.rename(columns=col_map).drop(columns=["time_years"])

# ── 3. Make Scenario_ID 1-based to match MVP1 convention ──────────────────────
hjm["Scenario_ID"] = hjm["Scenario_ID"] + 1

# ── 4. Sort scenario-major: all timesteps per scenario, then next scenario ─────
yield_cols = ["Y_1M", "Y_3M", "Y_6M", "Y_1Y", "Y_2Y",
              "Y_3Y", "Y_5Y", "Y_7Y", "Y_10Y", "Y_20Y", "Y_30Y"]

result = (
    hjm[["Scenario_ID", "Month"] + yield_cols]
    .sort_values(["Scenario_ID", "Month"])
    .reset_index(drop=True)
)

# ── 5. Save ────────────────────────────────────────────────────────────────────
result.to_csv("hjm_pca_reformatted.csv", index=False) # change filename as needed

print(f"Done. Shape: {result.shape}")
print(f"Scenarios : {result['Scenario_ID'].nunique()}")
print(f"Timesteps : {result['Month'].nunique()}  ({result['Month'].min()} to {result['Month'].max()})")
print(f"Columns   : {list(result.columns)}")
```
