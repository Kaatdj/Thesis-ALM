```python
import pandas as pd

# ── 1. Load data ──────────────────────────────────────────────────────────────
df = pd.read_csv("raw_data.csv")

# ── 2. Sample x random yield curves (rows) ───────────────────────────────────
x = 10                                         # ← adjust number of samples here
df_sampled = df.sample(n=x, random_state=42)   # remove random_state for truly random each run

# ── 3. Remove macro columns (GDP, CPI, Fed Funds rate) ───────────────────────
macro_cols = ["Real_GDP", "CPI_Level", "FedFunds"]
df_sampled = df_sampled.drop(columns=macro_cols)

# ── 4. Remove identifier columns (DATE + row index) ──────────────────────────
# Comment out this block if you want to keep DATE / index for reference
id_cols = ["DATE"]
df_sampled = df_sampled.drop(columns=id_cols)
df_sampled = df_sampled.reset_index(drop=True)  # drop the original row index too

# ── 5. Rename yield columns ──────────────────────────────────────────────────
df_sampled = df_sampled.rename(columns={
    "Y_DGS1MO":  "Yield_1M",
    "Y_DGS3MO":  "Yield_3M",
    "Y_DGS6MO":  "Yield_6M",
    "Y_DGS1":    "Yield_1Y",
    "Y_DGS2":    "Yield_2Y",
    "Y_DGS3":    "Yield_3Y",
    "Y_DGS5":    "Yield_5Y",
    "Y_DGS7":    "Yield_7Y",
    "Y_DGS10":   "Yield_10Y",
    "Y_DGS20":   "Yield_20Y",
    "Y_DGS30":   "Yield_30Y",
})

# ── 6. Save ───────────────────────────────────────────────────────────────────
df_sampled.to_csv("sampled_yield_curves.csv", index=False)
print(f"Saved {len(df_sampled)} sampled yield curves → sampled_yield_curves.csv")
print(df_sampled.head())
```
